```python
import matplotlib.pyplot as plt
import numpy as np

# データを作成
# 0から10までの幅で配列を等間隔に100個作成する
x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)

# プロット
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0)

ax.set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# データを作成
# 0から10までの幅で配列を等間隔に30個作成する
x = np.linspace(0, 10, 30)
y = 2 + 2 * np.sin(2 * x)

# プロット
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0)

ax.set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))

plt.show()
```

## 5.5 ヒストグラムを表示しよう

```python
import numpy as np
import matplotlib.pyplot as plt

#　平均が15, 標準偏差5の標準正規分布から2000個の乱数を生成する
x = np.random.normal(15, 5, 2000)

# ヒストグラムをプロット
plt.hist(x)
```

```python
import numpy as np
import matplotlib.pyplot as plt

#　平均が15, 標準偏差5の標準正規分布から2000個の乱数を生成する
x = np.random.normal(15, 5, 2000)

# ヒストグラムをプロット
plt.hist(x, bins=30)
```

## 箱ひげ図を表示
